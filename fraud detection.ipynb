#  Now, applying KNN on creditcard dataset

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.cross_validation import cross_val_score
from collections import Counter
from sklearn.metrics import accuracy_score
from sklearn import cross_validation
```

```python
#sampling our data
sampled_data=credit.sample(frac=1)
```

```python
#as KNN has a very large time complexity and we dont have so much we computation power so using 20k points instead of whole data is better

final_data=sampled_data[0:50000]
```

```python
#sorting our data with respect to time of each transaction 
final_data=final_data.sort_values('Time')
```

```python
X = final_data[['V3','V4','V9','V10','V11','V12','V14' ]]
y = final_data['Class']
```

```python
sampled_data=sampled_data.sort_values('Time')
X = sampled_data[['V3','V4','V9','V10','V11','V12','V14' ]]
y = sampled_data['Class']
X_train=X[0:256326]
X_test=X[256326:]
y_train=y[0:256326]
y_test=y[256326:]

myList = list(range(0,50))
neighbors = list(filter(lambda x: x % 2 != 0, myList))
cv_scores = []
# perform 10-fold cross validation
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())

# changing to misclassification error
MSE = [1 - x for x in cv_scores]

# determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print('\nThe optimal number of neighbors is %d.' % optimal_k)
```

```python
#knn_optimal = KNeighborsClassifier(n_neighbors=optimal_k)
# fitting the model
X = sampled_data[['V3','V4','V9','V10','V11','V12','V14' ]]
y = sampled_data['Class']
X_train=X[0:256326]
X_test=X[256326:]
y_train=y[0:256326]
y_test=y[256326:]
knn_optimal = KNeighborsClassifier(n_neighbors= optimal_k , weights='distance')

# fitting the model
knn_optimal.fit(X_train, y_train)

# predict the response
pred = knn_optimal.predict(X_test)

# evaluate accuracy
acc = accuracy_score(y_test, pred) * 100
print('\nThe accuracy of the knn classifier for k = %d is %f%%' % (optimal_k, acc))
```

# Performing logistic regression  

```python
#logistic regression without doing any hyper paramter tuning
sampled_data=sampled_data.sort_values('Time')
from sklearn.linear_model import LogisticRegression
X = sampled_data[['V3','V4','V9','V10','V11','V12','V14' ]]
y = sampled_data['Class']
X_train=X[0:256326]
X_test=X[256326:]
y_train=y[0:256326]
y_test=y[256326:]
lr = LogisticRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
acc = accuracy_score(y_test, pred) * 100
print('\nThe accuracy for logistic regression is %f%%' % (acc))
```

```python
#logistic regression with hyper parameter tuning using random search to improve accuracy



from sklearn.model_selection import train_test_split
from sklearn.grid_search import RandomizedSearchCV

from sklearn.datasets import *
from sklearn.linear_model import LogisticRegression



tuned_parameters = {'C': np.random.uniform(10**-4,10**4,1000)}



#Using GridSearchCV
model =  RandomizedSearchCV(LogisticRegression(class_weight={1:10,0:1}), tuned_parameters, scoring = 'accuracy', cv=5)


model.fit(X_train, y_train)
                        

 
print(model.best_estimator_)
print(model.score(X_test, y_test))
```

# Randomforest classifier

```python
from sklearn.ensemble import RandomForestClassifier


rf = RandomForestClassifier(n_estimators=10, max_depth=None,min_samples_split=2, random_state=10)
rf.fit(X_train,y_train)
scores = cross_val_score(rf, X_test, y_test)
scores.mean()
```

# Descision tree calssifier

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
scores = cross_val_score(clf, X_test, y_test)
scores.mean()
```

# SVM classifier

```python
from sklearn import svm
C = 1.0
svc = svm.SVC(kernel='sigmoid', C=C).fit(X_train, y_train)

scores = cross_val_score(svc, X_test, y_test)
scores.mean()
```

# Creating a stacking model using all the models used above


```python
def predict_(V3,V4,V9,V10,V11,V12,V14):
    
   # global pred,X_test
    from collections import Counter
    X_test = [[V3,V4,V9,V10,V11,V12,V14]]
    pred1 = model.predict(X_test)
    pred2= knn_optimal.predict(X_test)
    pred3 = rf.predict(X_test)
    pred4 = clf.predict(X_test)
    pred5= svc.predict(X_test)

    lst=[pred1[0],pred2[0],pred3[0],pred4[0],pred5[0]]
    

    data = Counter(lst)
    final_pred=data.most_common(1)[0][0]
    print(final_pred)
    return final_pred                                             #will return the final predicted value
```
